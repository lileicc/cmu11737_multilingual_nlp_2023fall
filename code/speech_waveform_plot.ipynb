```python
# imports
import matplotlib.pyplot as plt
import numpy as np
from pydub import AudioSegment

file_path = "record1.m4a"
audio_data = AudioSegment.from_file(file_path)
sample_rate = audio_data.frame_rate
audio_array = audio_data.get_array_of_samples()

time = np.linspace(
        0, # start
        len(audio_array) / sample_rate,
        num = len(audio_array)
    )

plt.figure(1, figsize=(16, 2))
plt.title("Sound Wave")
plt.xlabel("Time")
plt.plot(time, audio_array)
plt.savefig('data_waveform.png', transparent=True)
plt.show()

```
